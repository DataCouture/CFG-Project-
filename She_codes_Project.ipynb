```python
import csv

# ----Extended Project Idea 2.d: find out the months in which expenditure is more than sales
def months_expenditure_more_than_sales():
    months = []
    with open('sales.csv', newline='') as csv_file:
        reader = csv.reader(csv_file)
        next(reader)

        for row in reader:
            month = row[1]
            sales = float(row[2])
            expenditure = float(row[3])

            if expenditure > sales:
                months.append(month)

    return months

# ----Extended Project Idea 2.e: calculate profit after expenditure
def profit_after_expenditure(month_to_check):
    with open('sales.csv', newline='') as csv_file:
        reader = csv.reader(csv_file)
        next(reader)

        for row in reader:
            month = row[1]
            if month == month_to_check:
                sales = float(row[2])
                expenditure = float(row[3])
                profit_after_expenditure = sales - expenditure
                return profit_after_expenditure

    # If the month is not found in the data
    return "Data not available for the given month."


months_expenditure_more = months_expenditure_more_than_sales()
print("Months where expenditure is more than sales:", months_expenditure_more)


month_to_check = input ( "Month for Profit after expenditure" )
profit_after_expenditure = profit_after_expenditure(month_to_check)
if isinstance(profit_after_expenditure, float):
    print(f"The profit after expenditure for {month_to_check} is ${profit_after_expenditure:.2f}")
else:
    print(profit_after_expenditure)


```
